# Emotion-Aware Chatbot using LLM

## Project Overview
This project integrates sentiment analysis into a chatbot to detect 
Positive, Negative, or Neutral emotions and respond appropriately.

## Objectives
- Detect sentiment from user messages
- Generate sentiment-aware responses
- Improve customer interaction quality


```python
# Importing All Esstinal Libraries

import os
from dotenv import load_dotenv
from groq import Groq
```

We Imported Required Libraries For Environment Management And LLM Interaction.

```python
load_dotenv()
client = Groq(api_key=os.getenv("GROQ_API_KEY"))
```

The API Key Is Securely Loaded Using a `.env` file.<br>
We Must Not Share It With Anyone

```python
PROMPT = """
You Are a Customer Support Chatbot.

For every user message:
1. Classify Sentiment As Positive, Negative Or Neutral.
2. Respond Appropriately And According To The Sentiment.

Respond In This Format
Sentiment: <Positive | Negative | Neutral.
Response: <short professional reply>

User Message:
{user_message}
"""
```

This Prompt Controls How The LLM Performs Sentiment detection And Response Generation.

```python
def get_sentiment_response(user_input):
    final_prompt = PROMPT.format(user_message = user_input)

    response = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[{"role": "user", "content": final_prompt}]
    )

    return response.choices[0].message.content
```

This function sends user input to the LLM and returns sentiment-aware output.

```python
user_input = "I Find Your Product Good"
print(get_sentiment_response(user_input))
```

The chatbot correctly detects sentiment and generates an appropriate response.
