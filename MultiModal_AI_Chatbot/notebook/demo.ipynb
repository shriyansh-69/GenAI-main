# **Import Libraries**

```python
from google import genai
from dotenv import load_dotenv
from PIL import Image
import os
```

## **Load Environment Variable**

```python
load_dotenv()
```

## **Intialize Gemini Client** 

```python
client = genai.Client(api_key=os.getenv("GOOGLE_API_KEY"))
```

## **Basic Text Chatbot** 

```python
prompt = "Explain Artificial Intelligence In Simple Words"

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt
)

print(response.text)
```

## **Image Understanding**

```python
image = Image.open("sample.png")
image
```

## Analyze Image Using Gemini

```python
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=[
        "Describe This Image",
        image
    ]
)

print(response.text)
```

## Multimodal interaction

```python
question = "ExplainThe first Column"

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=[question,image]
)

print(response.text)
```

## **Create a Reusable Chat Function**

```python
def get_response(user_input):

    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=user_input
    )

    return response.text
```

## **Test Chat Function**

```python
answer = get_response("Explain Deep Learning")
print(answer)
```

## **Create Reusable Image Function**

```python
def process_image(user_input, image):

    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=[user_input,image]
    )


    return response.text
```

## **Test Image Function**

```python
image = Image.open("sample.png")

result = process_image(
    "Explain This Image",
    image)

print(result)
```
